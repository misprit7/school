```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.dpi'] = 150 # 200 e.g. is really fine, but slower
plt.rcParams.update({'font.size': 11})
```

```python
m1=1
m2=4
mu=m1*m2/(m1+m2)
l={1:2, 2:1, 3:6}
n=[1, 2, 3]
Gn={1:1, 2:1, 3:3}
alpha={i:m1*m2*Gn[i] for i in n}
```

```python
plt.rcParams['figure.figsize'] = [9, 3]
figure, axis = plt.subplots(1, 4)
# figure.set_figwidth(15)
# figure.set_figheight(7)
r=np.linspace(0.5, 4, 1000)
U=[l[i]**2/2/mu/r**2-alpha[i]/r**i for i in n]
axis[0].plot(r, U[0])
axis[1].plot(r, U[1])
axis[2].plot(r, -U[1])
axis[3].plot(r, U[2])

figure.tight_layout(rect=[0, 0.03, 1, 0.87])
figure.suptitle('$U(r)$ for Different Values of $n$')
axis[0].set_title('$n=1$')
axis[1].set_title('$n=2, C<0$')
axis[2].set_title('$n=2, C>0$')
axis[3].set_title('$n=3$')
plt.setp(axis[:], xlabel='$r$');
plt.setp(axis[0], ylabel='$U(r)$');

```

```python
l=2
mu=1
alpha2=1
E=2
C=l**2/2/mu-alpha2
e=np.sqrt(np.abs(1-2*mu*alpha2/l**2))
p2=np.sqrt(C/E)

phi1 = np.linspace(-1/2/e*np.pi+1e-2, 1/2/e*np.pi-1e-2, 10000)
r1=p2/np.cos(e*phi1)
phi1 = np.linspace(0, 1/2/e*np.pi-1e-2, 10000)
r3=p2/np.sinh(e*phi1)
phi2 = np.linspace(0, 2*np.pi, 10000)
r2=-1/np.sqrt(2*mu*E)/phi2
print(e, p2, C)
```

```python
plt.rcParams['figure.figsize'] = [8, 3]
figure, axis = plt.subplots(1, 4)

axis[0].plot(r1*np.cos(phi1), r1*np.sin(phi1))
axis[0].set_xlim(-4, 4)
axis[0].set_ylim(-4, 4)
axis[0].set_aspect('equal')


axis[1].plot(r3*np.cos(phi1), r3*np.sin(phi1))
axis[1].set_xlim(-4, 4)
axis[1].set_ylim(-4, 4)
axis[1].set_aspect('equal')

axis[2].plot(2*np.cos(phi2), 2*np.sin(phi2))
axis[2].set_xlim(-4, 4)
axis[2].set_ylim(-4, 4)
axis[2].set_aspect('equal')

axis[3].plot(8*r2*np.cos(phi2), 8*r2*np.sin(phi2))
axis[3].set_xlim(-4, 4)
axis[3].set_ylim(-4, 4)
axis[3].set_aspect('equal')

figure.tight_layout(rect=[0, 0.03, 1, 0.87])
figure.suptitle('Possible Trajectories for the $n=2$ Case')
axis[0].set_title(r'$C>0$')
axis[1].set_title(r'$C<0$')
axis[2].set_title(r'$C=E=0$')
axis[3].set_title(r'$C=0, E>0$')
plt.setp(axis[:], xlabel='x');
plt.setp(axis[0], ylabel='y');
```
