```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (10, 7)
```

```python
V0=1
a=2

delta = 0.025
x = np.arange(0, a, delta)
y = np.arange(0, a, delta)
X, Y = np.meshgrid(x, y)

V=sum(4*V0/n/np.pi/np.cosh(n*np.pi)*np.sin(np.pi*Y*n/a)*np.cosh(np.pi*n*X/a) for n in [2*i+1 for i in range(20)])
Ex, Ey=np.gradient(V)

plt.figure()
CS = plt.contour(X, Y, V)
plt.clabel(CS, inline=1, fontsize=10)
plt.title('Countour Plot of Potential')
plt.show()


fig, ax = plt.subplots()
q = ax.streamplot(X, Y, -Ey, -Ex)
plt.title('Electric Field for 1b')
plt.show()
```
