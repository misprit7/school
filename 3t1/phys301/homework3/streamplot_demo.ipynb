```python
import numpy as np
import matplotlib.pyplot as plt

# Grid of x, y points
L=4
nx, nz = 100, 100
x = np.linspace(-L, L, nx)
z = np.linspace(-L, L, nz)
X, Z = np.meshgrid(x, z)

a=1
b=2
q=1
k=9e9

charges = [(-q, (a, 0, 0)), (-q, (-a, 0, 0)), (3*q, (0, 0, b)), (-q, (0, 0, -b))]
E = [sum(-k*charge[0] / ((X-charge[1][0])**2+(Z-charge[1][2])**2)**(3/2) * (charge[1][2*i]-(X, Z)[i]) for charge in charges) for i in (0, 1)]


# demo of streamplot for a vector field
fig = plt.figure(figsize = (8,8))
plt.streamplot(x, z, E[0], E[1], linewidth=1, density=2, arrowstyle='->', arrowsize=1.5);
```
