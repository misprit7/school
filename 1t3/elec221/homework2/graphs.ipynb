```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
z=1+1j
z_conj=np.conj(z)
z_inv = 1/z
```

```python
plt.scatter(z.real, z.imag, label='z')
plt.scatter(z_conj.real, z_conj.imag, label='z conjugate')
plt.scatter(z_inv.real, z_inv.imag, label='z inverse')
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.legend()

plt.show()
```

```python
x=[2, 4, 6, 0, 2, 0]
h=[3, -1, 0, 1]
y=np.convolve(x, h)
plt.stem(range(len(y)), y)
```
