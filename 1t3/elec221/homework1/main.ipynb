```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 2b
fdelta = 20
fc = 200

# Init arrays
t = np.linspace(0, 100, 10000)
s = np.cos(2 * np.pi * fdelta * t) * np.cos(2 * np.pi * fc * t)
c = np.cos(40 * np.pi * t)

# Plot the data
plt.plot(t, s)
plt.plot(t, c)
plt.plot(t, -c)

plt.xlabel('Time (s)')
plt.ylabel('Signal output')
plt.title('Signal Output of Signals for 2b')

plt.show()
```

```python
from IPython.display import Audio
import numpy as np

sr = 10000
t = np.linspace(0, 2, sr * 2)
s = np.cos(np.pi * 2 * 20 * t) * np.cos(np.pi * 2 * 200 * t)
c = np.cos(2 * np.pi * 200 * t)
Audio(s, rate=sr)
Audio(c, rate=sr)
```
