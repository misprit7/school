```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.integrate as integrate
import control
```

```python
omega = np.linspace(-np.pi, np.pi, 100)
X = np.exp(-5*1j*omega)
fig, ax = plt.subplots(2, 1, constrained_layout=True)

ax[0].plot(omega, np.abs(X))
ax[1].plot(omega, np.angle(X))
fig.suptitle('Fourier transform for question 1a')
ax[0].set_title('Magnitude')
ax[1].set_title('Phase')
fig.savefig('images/q1a.png')
```

```python
omega = [np.pi/4, -np.pi/4, 5*np.pi/4, -5*np.pi/4]
plt.stem(omega, [2*np.pi, 2*np.pi, 4*np.pi, 4*np.pi])
plt.title('Magnitude of fourier transform of s(t)')
plt.xlabel('omega')
plt.savefig('images/q2e.png')
plt.show()
```

```python
omega = [5*np.pi/4, -5*np.pi/4]
plt.stem(omega, [4*np.pi, 4*np.pi])
plt.title('Magnitude of fourier transform of y(t)')
plt.xlabel('omega')
plt.savefig('images/q2g.png')
plt.show()
```

```python
omega = np.linspace(-np.pi, np.pi, 100)
X = np.array([1 if np.abs(w) < np.pi/2 else 0 for w in omega])
plt.plot(omega, X)
plt.title('Fourier Transform of x(t)')
plt.xlabel('omega')
plt.savefig('images/q3b.png')
plt.show()
```

```python
omega = np.linspace(-2*np.pi, 2*np.pi, 100)
Y = 1/2 * np.ones_like(omega)
plt.plot(omega, Y)
plt.title('Fourier transform of y(t)')
plt.xlabel('omega')
plt.savefig('images/q3c.png')
plt.show()
```

```python
t = np.linspace(-2*np.pi, 2*np.pi, 100)
y = [1/2 if np.abs(T) < 1e-1 else 0 for T in t]
plt.plot(t, y)
plt.title('y(t)')
plt.xlabel('t')
plt.savefig('images/q3d.png')
plt.show()
```

```python
t = np.linspace(-20, 20, 200)
y = [integrate.quad(lambda T: np.sin(np.pi*T/2)/(np.pi * T), -99, tau) for tau in t]

plt.plot(t, y)
plt.title('Step response for question 3f')
plt.xlabel('t')
plt.savefig('images/q3f.png')
plt.plot()
```

```python
t = np.linspace(0, 20, 200)
y = 1-np.exp(-t/2)

plt.plot(t, y)
plt.title('Step response for question 4b')
plt.xlabel('t')
plt.savefig('images/q4b.png')
plt.plot()
```

```python
omega = np.linspace(-np.pi, np.pi, 100)
X = 1 / (1 + 2j*omega)
fig, ax = plt.subplots(2, 1, constrained_layout=True)

ax[0].plot(omega, np.abs(X))
ax[1].plot(omega, np.angle(X))
fig.suptitle('Fourier transform for question 4c')
ax[0].set_title('Magnitude')
ax[1].set_title('Phase')
fig.savefig('images/q4c.png')
```
