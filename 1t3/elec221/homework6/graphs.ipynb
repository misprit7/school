```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
omega = np.linspace(-3*np.pi, 3*np.pi, 1000)
Xd = [1-np.abs(w%(2*np.pi))/np.pi for w in omega]

plt.plot(omega, Xd)
plt.savefig('images/2a.png')
plt.show()
```

```python
Yd = [1-np.abs(w/np.pi) if np.abs(w) <= np.pi/2 else 0 for w in omega]

plt.plot(omega, Yd)
plt.savefig('images/2b.png')
plt.show()
```

```python
omegap = np.linspace(-300, 300, 10000)
Yp = np.zeros(10000)
for n in range(-1, 2):
    Yp += np.array([1-np.abs(w/100-2*n) if np.abs(w/100-2*n) <= 1/2 else 0 for w in omegap])
Yp *= 100/np.pi
plt.plot(omegap, Yp)
plt.savefig('images/2c.png')
```

```python
Yc = [1-np.abs(w/100) if np.abs(w/100) <= 1/2 else 0 for w in omegap]

plt.plot(omegap, Yc)
plt.savefig('2d.png')
```
