```python
def moving_average(a, n=10):
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n

I = moving_average(io.loadmat('data/6B.mat')['contrast'].flatten()[40:-10])
# peaks, _ = scipy.signal.find_peaks(-I, prominence=30, height=-18)
# plt.plot(peaks/30, I[peaks], "x")
plt.plot(np.array(range(len(I)))/30*1.06/1e3+9.715, I)
# plt.plot(I)
plt.title('Contrast of Fringes of White Light With Filter')
plt.xlabel('Distance (mm)')
plt.ylabel('Contrast (pixel value)')
plt.show()
```
